# Uporabniško definirane funkcije

Funkcija je zaključen sklop programske kode, ki izvede določeno opravilo. Funkcije lahko kličemo iz različnih delov svojega programa, kadar želimo izvesti opravilo, ki ga izvaja funkcija. Funkcijo si lahko predstavljamo kot črno škatlo, v katero vstavimo vhodne podatke in ki vrne nabor rezultatov, ki so izhodni podatki funkcije. 

V Pythonu je na voljo veliko funkcij. Nekatere so del osnovnega Pythona, rečemo jim "vgrajene" (ang. _built-in_) in jih redno uporabljamo, kot na primer `print()`, `input()`, `len()`, `int()`, `str()`... 

Še več funkcij vsebujejo standardne knjižnice (ang. Python Standard Library). Funkcije, ki jih ne uporabljamo ravno v vsakem programu, so shranjene v standardnih knjižnicah, na primer knjižnici matematičnih funkcij __math__. Te funkcije je potrebno pred uporabo uvoziti z `import`. 

Vsak lahko ustvari tudi svoje funkcije, ki jih imenujemo uporabniško definirane funkcije. Tudi te so lahko shranjene v knjižnice. Primer so knjižnice, ki so jih naredili drugi, kot na primer knjižnica za risanje grafov __matplotlib__. Te knjižnice niso sestavni del osnovne, oziroma standardne namestitve Pythona. Da bi jih uporabljali, knjižnico najprej namestimo s programom __pip__ in nato iz knjižnice uvozimo funkcijo, kot v primeru standardne knjižnice. 

Nenazadnje funkcije lahko napišemo tudi sami. To poglavje je namenjeno prav temu.  

Funkcije so namenjene boljši organizaciji programske kode, olajšajo vzdrževanje in popravljanje programov ter omogočajo ponovno uporabo programske kode.

Novo funkcijo ustvarimo z rezervirano besedo `def`, ki ji sledi ime funkcije. Slednje si izmislimo in pri poimenovanju sledimo istim pravilom, kot pri poimenovanju spremenljivk. V oklepajih navedemo parametre, ki jih bo funkcija sprejela kot vhodne podatke. Telo funkcije, to so ukazi, ki jih funkcija ob klicu izvede, zamaknemo enako, kot v primeru zank ali pogojnih stavkov.

Da bo funkcija nekaj naredila, jo moramo v programu poklicati. Sicer je funkcija le definirana in če je ne pokličemo, se nikoli ne bo izvedla.

V nadaljevanju bomo pokazali, kako funkcijo definiramo in kako jo uporabimo. Osredotočili se bomo predvsem na prenos parametrov pri klicu funkcije in vračanju razultatov. Na koncu bomo pokazali, kako lahko ustvarimo svojo knjižnico funkcij.

>IZRAZI: 
>- V Pythonu namesto izraza _definicija_ pogosto uporabimo izraz _deklaracija_ funkcije. V besedilu bomo izraza uporabljali izmenjaje kot sopomenki. 
>- Pri definiciji funkcij določimo vhodne _parametre_. To so spremenljivke, ki bodo sprejele vhodne podatke. Funkcijo pokličemo tako, da ji posredujemo dejanske vrednosti parametrov. Tem vrednostim bomo v nadaljevanju rekli tudi _argumenti_. 

### Funkcija brez parametrov 

V prvem primeru poglejmo, kako lahko definiramo funkcijo. Naredili bomo funkcijo, ki ne sprejme nobenih argumentov in tudi nič ne vrne. 

```python
def moja_funkcija():
    print('Dober dan!')

#ko celico izvedemo, se funkcija ne izvede
```

Da bo funkcija v resnici nekaj naredila, jo moramo poklicati. 

>POZOR: \
>Pri definiciji in klicu funkcije moramo uporabiti oklepaje, tudi v primerih, ko funkciji ne posredujemo nobenih vhodnih podatkov.

```python
moja_funkcija()
```

### Vhodni in izhodni parametri 

S funkcijo običajno želimo opraviti določeno izračun ali transformacijo podatkov. Torej bo funkcija za delo potrebovala podatke. Podatke v funkcijo posredujemo v obliki vhodnih parametrov. Pri tem seveda ne definiramo funkcije s konkretnimi vrednostmi vhodnih podatkov, saj želimo funkcijo uporabiti za različne nabore podatkov. Namesto dejanskih vrednosti navedemo spremenljivke, v katerih bomo konkretne vrednosti posredovali ob klicu. Te spremenljivke imenujemo parametri funkcije.

Rezultate iz funkcije vračamo z ukazom `return`, za katerim navedemo vse spremenljivke, za katere želimo, da jih funkcija vrne. 

Definirajmo funkcijo, ki izračuna dolžino hipotenuze pravokotnega trikotnika, če poznamo dolžini obeh katet. Funkciji moramo posredovati dve števili, ki ju bo funkcija uporabila kot dolžini katet. Na koncu funkcija vrne eno število, ki predstavlja dolžino hipotenuze. 

```python
def hipotenuza(a, b):
    a_kvadrat = a**2
    b_kvadrat = b**2
    c = (a_kvadrat + b_kvadrat)**(1/2)
    return c

h = hipotenuza(3, 4)
print(h)
```

Iz primera vidimo, da parametre naštejemo v oklepajih med deklaracijo funkcije. Funkcijo pokličemo z argumenti, ki napolnijo parametre. Vrnjene vrednosti lahko shranimo v spremenljivke. Lahko jih tudi takoj uporabimo kot vhodne rednosti pri klicu drugih funkcij.

```python
print(hipotenuza(12, 9))
```

### Pozicijski parametri
V primeru zgoraj je vsebinsko vseeno kako naštejemo vrednosti, saj zamenjava dolžin katet ne vpliva na končni rezultat. Vendar pa v splošnem to ne velja in je pomembno, kateri parameter dobi katero vrednost.

Definirajmo funkcijo potenca, ki sprejme dva vhodna parametra, prvi je število, ki ga potenciramo, torej osnova in drugi je eksponent.

```python
def potenca(osnova, eksponent):
    return osnova**eksponent

potenca(4, 3)
```

Če bi vrstni red argumentov ob klicu funkcije obrnili, bi seveda dobili drugačen rezultat.

```python
potenca(3, 4)
```

Iz prikazanega vidimo, da funkcija sprejme argumente in napolni parametre v tistem vrstnem redu, kot smo jih navedli v oklepaju ob deklaraciji funkcije.

Privzeti način uporabe parametrov je torej odvisen od njihove pozicije med deklaracijo funkcije in zato te parametre imenujemo __pozicijski parametri__.

### Imenski parametri

Klic funkcije v katerem upoštevamo pozicijo parametrov je krajši, vendar moramo paziti na pravilen vrstni red in pri velikem številu parametrov lahko klici postanejo nepregledni.

Python omogoča še klic funkcije z imenskimi (ang. _keyword_) parametri. Deklaracija funkcije se pri tem nič ne spremeni, spremeni pa se način klicanja funkcije. Pri klicu lahko namreč navedemo ne le argument, temveč tudi ime parametra.

```python
potenca(eksponent=4, osnova=3)
```

Pri takšnem klicu funkcije ni več važno v katerem vrstnem redu naštejemo vrednosti, saj eksplicitno navedemo v katere parametre naj se argumenti zapišejo. 

>POMEMBNO: \
>V obeh primerih klicev, tako pozicijsko kot imensko, moramo navesti toliko vrednosti, kot je parametrov! 

### Privzeta vrednost argumenta

Kadar za določene parametre v naprej poznamo vrednosti, ki so običajne, oziroma želimo, da bi določene parametre ob klicu funkcije lahko tudi izpustili, lahko vrednosti teh pametrov določimo že ob deklaraciji funkcije. Nato te parametre lahko ob klicu funkcije spremenimo, jih zamenjamo z novimi argumenti, ali pa jih pri klicu enostavno izpustimo, saj že imajo določene svoje privzete vrednosti.  

Zapišimo funkcijo za izračun hipotenuze, ki predpostavlja, da je kateta $b$ dolžine 10.

```python
def hipotenuza_10(a, b=10):
    return (a**2 + b**2)**(1/2)

c = hipotenuza_10(5)
print(c)

c = hipotenuza_10(5, 20)
print(c)
```

#### Primer
Definirajmo funkcijo izpis(), ki ima dva vhodna parametra $a$ in $b$. Privzeto določimo $a=3$ in $b=5$. 

Primeri klicev funkcije prikazujejo različne možnosti uporabe.  

```python
def izpis(a = 3, b = 5):
    print(f'a = {a}')
    print(f'b = {b}')
    print('-------')

#uporabimo privzete argumente    
izpis()   

#podamo argumente pozicijsko, nepoimenovano
izpis(4, 6)

#podamo samo a
izpis(7)

#če podamo samo b, moramo uporabiti ime parametra
izpis(b=7)
```

### Poljubno število argumentov

Do sedaj smo vse definicije funkcij zapisali tako, da smo v naprej določili število parametrov. Če parametrom določimo privzete vrednosti, sicer ni potrebno vseh navesti ob klicu funkcije, vendar pa je njihovo število še vedno v naprej določeno.

Z uporabo podatkovne strukture n-terica lahko definiramo funkcijo tako, da en parameter sprejme poljubno število argumentov. Seveda moramo pri obdelavi podatkov nato ustrezno poskrbeti za to, da v naprej ne vemo, koliko argumentov bo podanih ob klicu funkcije.  

Parameter, ki sprejme več vrednosti, označimo z __*__.

Oglejmo si dva primera. Najprej napišimo funkcijo z enim argumentom v katerega bomo posredovali iterator, ki lahko vsebuje več elementov.

```python
def vsota(a):
    c = 0
    for i in a:
      c += i
    return c

vsota([1, 2, 3])  # a = [1, 2, 3] - kot argument posredujemo seznam
```

```python
vsota((1, 2, 3)) # a = (1, 2, 3) - kot argument posredujemo n-terico
```

V obeh primerih smo morali uporabiti oklepaje s katerimi smo elemente združili v kompleksnejšo podatkovno strukturo.

Funkcijo lahko definiramo tudi nekoliko drugače. Parameter $a$ bo po klicu vseboval vsa navedena števila kot n-terico.

```python
def vsota(*a):
    c = 0
    for i in a:
      c += i
    return c

vsota(1, 2, 3)  # a = (1, 2, 3)  vsi argumenti se samodejno združijo v n-terico 
```

### Izhodni parametri

```python

#izhodni parametri, en, dva in več
def vsota_in_produkt(a, b):
    v = a + b
    p = a * b
    return v, p

v, p = vsota_in_produkt(2, 3)
```

### Doseg spremenljivke

Funkcije ustvarijo svoj naslovni prostor za spremenljivke, kar pomeni, da spremenljivke, ki jih uporabimo v funkciji, ne moremo uporabiti zunaj funkcije, saj po povratku iz funkcije več ne obstajajo. 

V spodnjem primeru znotraj funkcije uporabljamo spremenljivko $a$. Vendar tudi zunaj funkcije uporabljamo spremenljivko $a$. Pomembno je vedeti, da sta to dve različni spremenljvki, čeprav sta poimenovani enako. Spremenljivka znotraj funkcije je __lokalna__ spremenljivka in tista zunaj je __globalna__ spremenljivka. V Pythonu brez posebne deklaracije v funkciji ni mogoče spreminjati globalnih spremenljivk. Prav tako pa lokalne spremnljivke zunaj funkcije niso več dostopne. 

```python
def spremeni(x):
    a = x
    print(f'a znotraj funkcije spremeni: {a}')

a = 5
spremeni(3)
print(f'a izven funkcije spremeni: {a}')
```

Če bi želeli znotraj funkcije spreminjati globalne spremenljivke, moramo to napovedati. To naredimo z ukazom `global`. Na ta način omogočimo, da ima funkcija stranske učinke, saj posega v svoje okolje. Praviloma se takšnim posegom izogibamo. Stranski učinki so nepredvidljivi, saj v času, ko funkcijo napišemo, nikoli ne vemo, kje vse bo funkcija uporabljena v prihodnje. Uporabniki funkcije pa tudi ne morejo vedeti, da s klicem funkcije lahko "dobijo" več, kot so želeli.  

```python
def spremeni(x):
    global a
    a = x
    print(f'a znotraj funkcije spremeni: {a}')

a = 5
spremeni(3)
print(f'a izven funkcije spremeni: {a}')
```

Z ukazom `global` želimo povezati ali pripeljati neko globalno spremenljivko v funkcijo. Če ta spremenljivka zunaj funkcije še ne obstaja, bi to lahko predstavljalo težavo. Zato ta ukaz spremenljivko ne samo poveže, temveč po potrebi tudi ustvari. 

```python
def spremeni(x):
    global a
    a = x
    print(f'a znotraj funkcije spremeni: {a}')

spremeni(3)
print(f'a izven funkcije spremeni: {a}')
```

Spremenljivo ustvarimo tako, da ji priredimo vrednost. Če spremenljivki ne priredimo vrednosti, torej takšna spremenljivka ne obstaja in je ne moremo uporabiti. 

V spodnjem primeru bi lahko pričakovali, da bo klic funkcije `izpis` javil napako, saj uporabljamo spremenljivko $a$, ki je v funkciji nismo ustvarili. Vendar če pogledamo nižje, vidimo, da spremenljivko $a$ ustvarimo pred klicem funkcije. Torej $a$ obstaja kot globalna spremenljivka. V Pythonu bo funkcija globalno spremenljivko znala uporabiti in do napake ne bo prišlo.

Poudariti je potrebno, da se tudi takšni rabi globalne spremenljivke znotraj funkcije izogibamo, čeprav funkcija ne bo ustvarjala stranskih učinkov, saj ne vpliva na vrednost spremenljivke. Problem nastane, če pred klicem funkcije `izpis()` spremenljivke $a$ ne bi ustvarili. V tem primeru bo program javil napako, da uporabljamo spremenljivko, ki ne obstaja. Namig: poskusite spremeniti print(a) v print(z) in izvesti program. S takšno prakso ustvarjamo funkcije, ki so odvisne od okolja uporabe, česar pa ne želimo.   

```python
#uporabi globalno, če ni lokalne, sicer lokalno
def izpis():
    print(a)

a = 5
print(a)
izpis()
```

### Prenos parametrov
Funkcije za delovanje praviloma potrebujejo podatke. Ustrezen način posredovanja podatkov je uporaba parametrov funkcije. V Pythonu velja, da se parametri vedno prenašajo __po vrednosti__. To pomeni, da so parametri dejansko lokalne spremenljivke funkcije. Če funkcijo pokličemo z argumenti, ki so spremenljivke, se vrednosti teh spremenljivk prekopirajo v parametre. To velja za vse enostavne podatkovne tipe. 

Drugače je s sestavljenimi podatkovnimi tipi. Če bomo znotraj funkcije spreminjali elemente sestavljenih podatkovnih tipov (npr. elemente seznama), se bo to odražalo tudi izven funkcije.  

V spodnjem primeru funkciji `spremeni()` kot argument posredujemo spremenljivko $a$. Vrednost $a$ se bo prekopirala v lokalno spremenljivko $x$. Karkoli bomo z $x$ v funkciji počeli, to ne bo imelo vpliva na vrednost $a$.

```python
def spremeni(x):
    print(f'a v funkciji pred spremembo: {x}')
    x = 77
    print(f'a v funkciji po spremembi: {x}')

a = 5
spremeni(a)
print(f'a izven funkcije spremeni: {a}')
```

Naj bo $a$ spremenljivka sestavljenega podatkovnega tipa `list` (seznam). Če kateri element seznama znotraj funkcije spremenimo, se bo to odražalo tudi zunaj funkcije, ko le-ta zaključi svoje delovanje. 

```python
def spremeni(x):
    x[1] = 77
    print(f'a v funkciji spremeni: {x}')
    
a = [1, 2, 3]
spremeni(a)
print(f'a v funkciji spremeni: {a}')
```

>POMEMBNO: \
>Dobra praksa je, da parametre, ki jih navedemo v oklepajih pri definiciji funkcije, vedno uporabimo kot VHODNE spremenljivke, torej za prenos vrednosti, ki jih v funkcijo želimo posredovati. Za prenos rezultatov pa uporabimo `return` s katerim iz funkcije vrnemo vse, kar od funkcije pričakujemo. IZHODNE spremenljivke torej posredujemo z ukazom `return`.
