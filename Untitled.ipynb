```python
def g(x):
    return 1/x

I = quad(g, 1e-9, np.inf, weight='sin', wvar=1)

I
```

```python
np.pi/2 - I[0]
```

```python
def f(x):
    return np.exp(-x)/np.sqrt(x)

I = quad(f, 0, np.inf)
I
```

```python
def f(x, a, b):
    return (np.sin(b*x)/x) * np.exp(-a*x)
```

```python
I1 = quad(f, 0, np.inf, args=(1, 20))
I2 = quad(f, 0, np.inf, args=(20, 20))

I1, I2
```

```python
sez_a = np.arange(1, 10)
sez_b = np.arange(1, 10)

I_ab = [[a, b, quad(f, 0, np.inf, args=(a, b))[0]] for a in sez_a for b in sez_b]
I_ab
```

```python
def f(x):
    return np.exp(-(x-300)**2) + np.exp(-(x+300)**2)
```

```python
fig, ax = plt.subplots()
x = np.linspace(-400, 400, 2000)
y = f(x)
ax.set_xlim(290, 310)
ax.plot(x,y)
```

```python
quad(f, -10000, 10000, points=[-300, 300])
```

```python
poraba = (-np.arange(-500, 500, 10)**2)
poraba = poraba - min(poraba)
poraba = poraba - 40000 * np.random.rand(100)

fig, ax = plt.subplots()
dnevi = range(1, 101)
ax.scatter(dnevi, poraba);
```

```python
integral = cumulative_trapezoid(poraba, dnevi, initial=0)
int2 = trapezoid(poraba, dnevi)

fig, ax = plt.subplots()
dnevi = range(1, 101)
ax.plot(dnevi, integral);
int2, integral[-1]
```
